#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("sinasc_RO_2019.csv")
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Variáveis a serem analisadas
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

# Criando gráficos de barras para cada variável
for variavel in variaveis:
    plt.figure(figsize=(10, 6))
    df[variavel].value_counts().sort_index().plot(kind='bar', color='skyblue')
    plt.title(f'Gráfico de Barras para {variavel}')
    plt.xlabel(variavel)
    plt.ylabel('Contagem')
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Calculando a média do APGAR5 por CONSULTAS
media_apgar5_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Criando o gráfico de barras
plt.figure(figsize=(12, 6))
media_apgar5_por_consultas.plot(kind='bar', color='skyblue')
plt.title('Média do APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média do APGAR5')
plt.xticks(rotation=45, ha='right')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Definindo o layout do painel
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico de barras para SEXO
sns.countplot(data=df, x='SEXO', palette='pastel', ax=axs[0, 0])
axs[0, 0].set_title('Distribuição de SEXO')

# Gráfico de barras para GESTACAO
sns.countplot(data=df, x='GESTACAO', palette='pastel', ax=axs[0, 1])
axs[0, 1].set_title('Distribuição de GESTACAO')

# Gráfico de barras para ESTCIVMAE
sns.countplot(data=df, x='ESTCIVMAE', palette='pastel', ax=axs[1, 0])
axs[1, 0].set_title('Distribuição de ESTCIVMAE')

# Gráfico de barras para CONSULTAS
sns.countplot(data=df, x='CONSULTAS', palette='pastel', ax=axs[1, 1])
axs[1, 1].set_title('Distribuição de CONSULTAS')

# Ajustando o layout
plt.tight_layout()
plt.show()
```
